```python
import altair as alt
from evaluator import evaluate
from daphne import daphne
import numpy as np
import pandas as pd
alt.data_transformers.disable_max_rows()
import sys
sys.setrecursionlimit(10000)
```

```python
ast_1 = daphne(['desugar-hoppl', '-i', '/Users/aliseyfi/Documents/UBC/Semester3/Probabilistic-Programming/HW/Probabilistic-Programming/Assignment_5/programs/{}.daphne'.format(1)])
print('\n\n\nSample of prior of program {}:'.format(1))
samples_1 = np.zeros(1000)
for i in range(1000):
    samples_1[i] = evaluate(ast_1)
```

```python
df_1 = pd.DataFrame({"Samples":samples_1})
alt.Chart(df_1).mark_bar().encode(
    alt.X("Samples:Q", bin=alt.Bin(maxbins=50)),
    y = 'count()'
)
```

```python
# marginal expectation
me_1 = np.mean(samples_1)
print(me_1)
```

```python
ast_3 = daphne(['desugar-hoppl', '-i', '/Users/aliseyfi/Documents/UBC/Semester3/Probabilistic-Programming/HW/Probabilistic-Programming/Assignment_5/programs/{}.daphne'.format(3)])
print('\n\n\nSample of prior of program {}:'.format(3))
samples_3 = np.zeros((17,1000))
for i in range(1000):
    samples_3[:,i] = evaluate(ast_3)
```

```python
df_3 = pd.DataFrame(samples_3)
```

```python
charts_1 = []
charts_2 = []
charts_3 = []
for i in range(17):
    df_tmp = pd.DataFrame({"step_{}".format(i):samples_3[i,:]})
    if i//6 == 0:
        charts = charts_1
    elif i//6 == 1:
        charts = charts_2
    else:
        charts = charts_3
    charts.append(alt.Chart(df_tmp).mark_bar().encode(
        alt.X("step_{}:Q".format(i), bin=True),
        y = 'count()'
    ))
tmp = []
tmp.append(alt.hconcat(*charts_1))
tmp.append(alt.hconcat(*charts_2))
tmp.append(alt.hconcat(*charts_3))

alt.vconcat(*tmp)
```
